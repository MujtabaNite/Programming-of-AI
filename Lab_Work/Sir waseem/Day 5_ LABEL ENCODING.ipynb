# **MORE ON ENCODING**


# **Type of Encoding**
 * ONEHOTENCODER  (vector making and order doesn't matter)
 * ORDINALENCODER  (as the name suggest the order here order matter)
 * LABELENCODER ()
 * FREQUENCY (time a value appear)
 * TARGET   (mean encoding)


```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder

import pandas as pd

```

```python
# sample data set with temperature and taste

temp_seq = ['hot','cold','warm']
taste_seq = ['sweet','sour','bitter']

data = {
    'temperature': ['hot', 'cold', 'warm', 'hot', 'cold'],
    'taste': ['sweet', 'sour', 'bitter', 'sweet', 'sour'],
    'target': ['Good', 'Bad', 'Good', 'Bad', 'Good']
}
```

```python
df = pd.DataFrame(data)
print("Original DataFrame:")
df
```

# **USING ORDINAL ENCODER**

```python
ordinal_encoder = OrdinalEncoder(categories=[temp_seq, taste_seq])  # in the form of list the cat are been passed here 

# the above line will store the seq and use the order of the sq from teh index of the list given



encoded_data = ordinal_encoder.fit_transform(df[['temperature', 'taste']])

print(f"the encoded data ")
encoded_data
```

# **USING LABEL ENCODER**

```python
label_encoder = LabelEncoder()

encoded_LABEL_data = label_encoder.fit_transform(df['target'])

print("the encoded data using the label encoder")

encoded_LABEL_data
```

# **USING THE ONE HOT ENCODER**

```python
Hot_Encoder = OneHotEncoder()

encoded_hot_encoder = Hot_Encoder.fit_transform(df)


print("the encoded data using the one hot encoder")

print(encoded_hot_encoder)

encoded_hot_encoder.toarray()


```
