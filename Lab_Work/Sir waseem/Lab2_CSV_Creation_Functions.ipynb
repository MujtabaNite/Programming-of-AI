# **CSV Creation and Function**

# Task to be done
  * Create a Series for the monthly expense
  * Handle Missing Values by filling them with 0 
  * Sort the series by Month in chronological order
  * Find the month with highest expense
  * Calculate teh total expenses across all months


```python
# Reading the csv and visualizing the data
import pandas as pd
data = pd.read_csv('MonthlyExpenses.csv')
data.head(3)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.set_theme(style="whitegrid")
sns.countplot(x='expensesCategory', data=data, palette="Set2")
plt.xlabel('Expenses Category')
plt.ylabel('Amount')
plt.title('Monthly Expenses')
plt.xticks(rotation=45)
plt.show()
```

```python
# Creating the new series because the monthly expenses are not in the  data set for that we have to create a new series on month and expensesCategory

data['MonthlyExpenses'] = data['expensesCategory'].map(data.groupby('expensesCategory')['Amount'].sum())

data.head(3)
```

```python
# handling missing values by filling them with 0
data.fillna(0, inplace=True)
data.head()
```

```python
# Sorting the data in ascending order by months by making a list of months and using it to sort the data
# we have to sort the data like first will be january, february, march and so on
# we can do this by making a list of months and using it to sort the data
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
data['Months'] = pd.Categorical(data['Months'], categories=months, ordered=True)
sorted_data = data.sort_values('Months')
data.head(3)

```

```python
# Calculating the total expense of the data
data_series_amount = pd.Series(data['Amount']).sort_values(ascending=True)
# Displaying the total expense
total_expense = data_series_amount.sum()
print("Total Expense: ", total_expense)
```
