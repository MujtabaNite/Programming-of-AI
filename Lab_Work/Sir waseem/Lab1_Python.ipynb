# implement example of dictionary 

```python
person = {
    'Name': 'Ali',
    'Age': 25,
    'Gender': 'M',
    'Occupation': 'Engineer'
}


print(person)
print("Name:", person['Name'])
print("Age:", person['Age'])
person['City'] = 'Karachi'
print(person)
```

```python
cars = {
    'Toyota': 'Corolla',
    'Honda': 'Civic',
    'Ford': 'Mustang',
    'Chevrolet': 'Camaro',
    'BMW': 'M3'
}

for brand, model in cars.items():
    print(f"Brand: {brand}, Model: {model}")
```

# Exploring Builtin Funcitons , properties ,methods

```python
car_dataset = {
    "cars": ['Audi', 'BMW', 'Chevrolet', 'Ford', 'Honda', 'Hyundai', 'Kia', 'Mazda', 'Mercedes-Benz', 'Nissan', 'Subaru', 'Toyota', 'Volkswagen'],
    "colors": ['Black', 'Blue', 'Brown', 'Gray', 'Green', 'Orange', 'Red', 'Silver', 'White', 'Yellow'],
    "years": ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019'],
    "prices": ['10000', '20000', '30000', '40000', '50000', '60000', '70000', '80000', '90000', '100000'],
    "models": [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
}

def Functions_dict(car_dataset):
    print("Cars:", car_dataset["cars"])
    print("Colors:", car_dataset["colors"])
    print("Years:", car_dataset["years"])
    print("Prices:", car_dataset["prices"])
    print("Models:", car_dataset["models"])
    print("Length of cars:", len(car_dataset["cars"]))
    print("Length of colors:", len(car_dataset["colors"]))
    print("Length of years:", len(car_dataset["years"]))
    print("Length of prices:", len(car_dataset["prices"]))
    print("Length of models:", len(car_dataset["models"]))
    print("Max of models:", max(car_dataset["models"]))
    print("Min of models:", min(car_dataset["models"]))
    print("Sum of models:", sum(car_dataset["models"]))
    print("Sorted models:", sorted(car_dataset["models"]))
    print("Reversed models:", reversed(car_dataset["models"]))
    print("Enumerated models:", enumerate(car_dataset["models"]))
    print("Zipped models:", zip(car_dataset["models"]))
    print("All models:", all(car_dataset["models"]))
    print("Any models:", any(car_dataset["models"]))
    print("List of models:", list(car_dataset["models"]))
    print("Tuple of models:", tuple(car_dataset["models"]))
    print("Set of models:", set(car_dataset["models"]))
    # print("Dictionary of models:", dict(car_dataset["models"]))
    print("copy of models:", car_dataset.copy())
    print("items of models:", car_dataset.items())
    print("keys of models:", car_dataset.keys())
    print("values of models:", car_dataset.values())
    print("pop of models:", car_dataset.pop("models"))
    print("popitem of models:", car_dataset.popitem())
    print("update of models:", car_dataset.update())
    print("setdefault of models:", car_dataset.setdefault("colors",""))
    print("fromkeys of models:", car_dataset.fromkeys("cars"))
    print("clear of models:", car_dataset.clear())



Functions_dict(car_dataset)



```
