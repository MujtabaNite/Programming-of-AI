```python
from sklearn.model_selection import train_test_split
import pandas as pd

# Example dataset
data = {'Feature': [1, 2, 3, 4, 5, 6], 'Target': [0, 1, 0, 1, 0, 1]}
df = pd.DataFrame(data)

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(
    df[['Feature']], df['Target'], test_size=0.33, stratify=df['Target'], random_state=42
)

print("Training target distribution:\n", y_train.value_counts())
print("Testing target distribution:\n", y_test.value_counts())
```

# Notes
   # The implementation is designed to:

* Generate test sets such that all contain the same distribution of classes, or as close as possible.
* Be invariant to class label: relabelling y = ["Happy", "Sad"] to y = [1, 0] should not change the indices generated.
* Preserve order dependencies in the dataset ordering, when shuffle=False: all samples from class k in some test set were contiguous in y, or separated in y by samples from   classes other than k.
* Generate test sets where the smallest and largest differ by at most one sample.


```python
from sklearn.model_selection import StratifiedKFold
from sklearn.datasets import make_classification

# Generate a dataset
X, y = make_classification(n_samples=500, n_classes=2, weights=[0.7, 0.3], random_state=42)

# Stratified K-Fold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)



for train_index, test_index in skf.split(X, y):
    print("Train class distribution:", y[train_index].mean())
    print("Test class distribution:", y[test_index].mean())
```

# **MinMax Scaling , Z-score**

```python
import numpy as np
import pandas as pd

def minmax(l):
    return [(x-min(l))/(max(l)-min(l))  for x in l]

def z_score(l):
    return pd.DataFrame([((x - np.mean(np.array(l))) / np.std(np.array(l)))  for x in l],columns=['z-score'])


# from sklearn.preprocessing import MinMaxScaler  # hamra acha hai

# from sklearn.preprocessing


print(np.std(np.array([20,55,15,27,45])))
print(np.mean(np.array([20,55,15,27,45])))

print(f"MinMAX  {minmax([20,55,15,27,45])}")
print(z_score([20,55,15,27,45]))
# print(MinMaxScaler([20,55,15,27,45]))

```
