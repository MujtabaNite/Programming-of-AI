```python
# importing lib

from imblearn.under_sampling import RandomUnderSampler 
# from imblearn import oversampling
from imblearn.over_sampling import RandomOverSampler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt







# Load the dataset
df = pd.read_csv('depression_dataset.csv')

# Display the first few rows of the dataset
print(df.head())

print(df.info())

# APPLYING oversampling and undersampling

df['depression'] = df['depression'].map({0: 'No', 1: 'Yes'})

# Separate features and target variable
X = df.drop('depression', axis=1)  # Features
y = df['depression']  # Target variable



# APPLYING oversampling and undersampling
undersampling = RandomUnderSampler(sampling_strategy='majority')
oversampling =  RandomOverSampler(sampling_strategy='minority')

x_resampled, y_resampled = undersampling.fit_resample(X, y)
X_resampled, Y_resampled = oversampling.fit_resample(X, y)

# Display the class distribution after resampling
print("Class distribution after resampling:")
print(y_resampled.value_counts())

print("Class distribution after resampling:")
print(Y_resampled.value_counts())


```

```python
# using of smote

import pandas as pd
from imblearn.over_sampling import SMOTE
from collections import Counter

# default value of k in smote is 5  we can change it to 1 or 2 or 3 or 4 or 5 for small dataet

X = pd.DataFrame({
    'X1': [1, 2, 3, 4, 5, 6],
    'X2': [1, 2, 3, 4, 5, 6],
    'X3': [1, 2, 3, 4, 5, 6],
})

y = [0, 1, 0, 1, 1, 1]

print("Original dataset shape %s" % Counter(y))
smote = SMOTE(random_state=42 , k_neighbors=1)
X_resampled, y_resampled = smote.fit_resample(X, y)
print("Resampled dataset shape %s" % Counter(y_resampled))
print(X_resampled)
print(y_resampled)  
```
