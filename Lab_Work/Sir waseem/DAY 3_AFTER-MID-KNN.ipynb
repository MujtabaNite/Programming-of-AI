# SIMPLE IMPUTER

```python
from sklearn.impute import SimpleImputer
import pandas as pd


simple =SimpleImputer(strategy="mean")
data = pd.read_csv(r"D:\University\University Work\Semestor 4\Programming of ai\LAB\Datasets\titanic_toy.csv")

```

```python
imputed_data = pd.DataFrame(simple.fit_transform(data))  # converting the data and the simple imputer to the data given 
```

# KNN IMPUTER

```python
from sklearn.impute import KNNImputer
import pandas as pd


KNNimputer = KNNImputer(n_neighbors=2)
KNN_IMPUTED_DATA = pd.DataFrame(KNNimputer.fit_transform(data))

KNN_IMPUTED_DATA.head()

```
