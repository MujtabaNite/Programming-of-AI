```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Task 1: Bar Chart (Simple & Grouped)
months = ["Jan", "Feb", "Mar"]
apples = [120, 145, 160]
bananas = [90, 110, 95]
oranges = [150, 130, 170]
```

```python
# Simple Bar Chart (Total Sales per Product)
products = ["Apples", "Bananas", "Oranges"]
total_sales = [sum(apples), sum(bananas), sum(oranges)]

plt.figure(figsize=(6, 4))
plt.bar(products, total_sales, color=["red", "yellow", "orange"])
plt.xlabel("Products")
plt.ylabel("Total Sales")
plt.title("Total Sales per Product")
plt.savefig("bar_chart_total_sales.png")
plt.show()
```

```python
# Grouped Bar Chart (Monthly Sales Comparison)
x = np.arange(len(months))
width = 0.2

plt.figure(figsize=(7, 5))
plt.bar(x - width, apples, width, label="Apples", color="red")
plt.bar(x, bananas, width, label="Bananas", color="yellow")
plt.bar(x + width, oranges, width, label="Oranges", color="orange")

plt.xticks(x, months)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales of Products")
plt.legend()
plt.savefig("bar_chart_grouped.png")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data (modify as needed)
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temp = [25, 27, 23, 29, 28, 26, 22]

# Convert days to numerical indices for plotting
day_indices = range(len(days))

plt.figure(figsize=(8, 5))
plt.plot(day_indices, temp, marker="o", label="Temperature")

# Annotate hottest and coldest days
max_temp = max(temp)
min_temp = min(temp)
max_day_index = temp.index(max_temp)
min_day_index = temp.index(min_temp)

plt.annotate(f"Hottest ({max_temp}°C)", 
             xy=(max_day_index, max_temp), 
             xytext=(max_day_index, max_temp + 1), 
             arrowprops=dict(facecolor="red", arrowstyle="->"))

plt.annotate(f"Coldest ({min_temp}°C)", 
             xy=(min_day_index, min_temp), 
             xytext=(min_day_index, min_temp - 1), 
             arrowprops=dict(facecolor="blue", arrowstyle="->"))

# Set x-axis labels correctly
plt.xticks(day_indices, days)

plt.xlabel("Days")
plt.ylabel("Temperature (°C)")
plt.title("Temperature Trend Over a Week")
plt.legend()
plt.grid()

# Save and show the plot
plt.savefig("line_chart_temperature.png")
plt.show()
```

```python
# Task 3: Stacked Bar Chart (Composition Analysis)
quarters = ["Q1", "Q2", "Q3", "Q4"]
marketing = [40, 45, 50, 55]
sales = [60, 70, 65, 80]
it = [20, 25, 30, 35]
```

```python
plt.figure(figsize=(7, 5))
plt.bar(quarters, marketing, label="Marketing", color="green")
plt.bar(quarters, sales, bottom=marketing, label="Sales", color="blue")
plt.bar(quarters, it, bottom=np.array(marketing) + np.array(sales), label="IT", color="purple")

plt.xlabel("Quarters")
plt.ylabel("Revenue")
plt.title("Quarterly Revenue Composition")
plt.legend()
plt.savefig("stacked_bar_chart.png")
plt.show()
```
