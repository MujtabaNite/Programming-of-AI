# **NEAR MISS ALL VERSION > HANDLING MISSING VALUES**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv(r"D:\University\University Work\Semestor 4\Programming of ai\LAB\Datasets\titanic_toy.csv")

sns.heatmap(data.isnull(),cmap='Oranges')
plt.show()
```

```python
data1 = pd.read_csv(r"D:\University\University Work\Semestor 4\Programming of ai\LAB\Datasets\data.csv")

sns.heatmap(data1.isnull(),cmap='Oranges')
plt.show()
```

# label encoding

```python
from sklearn.preprocessing import LabelEncoder

for col in data.columns:
    if data[col].dtype == 'object':
        le = LabelEncoder()
        data[col] = le.fit_transform(data[col])

for col in data1.columns:
    if data1[col].dtype == 'object':
        le = LabelEncoder()
        data1[col] = le.fit_transform(data1[col])

```

# APPLYING THE MISSING VALUE TECHNIQUE

# End of Distribution

```python
data.fillna((data.mean() + 3*data.std()), inplace=True)
data1.fillna((data.mean() + 3*data.std()), inplace=True)
```

# KNN

```python
from sklearn.impute import KNNImputer
import pandas as pd


KNNimputer1 = KNNImputer(n_neighbors=2)
KNN_IMPUTED_DATA1 = pd.DataFrame(KNNimputer1.fit_transform(data))

KNN_IMPUTED_DATA1.head()

```

```python
sns.heatmap(KNN_IMPUTED_DATA1.isnull(),cmap='Oranges')
plt.show()
```

```python

KNNimputer2 = KNNImputer(n_neighbors=2)
KNN_IMPUTED_DATA2 = pd.DataFrame(KNNimputer2.fit_transform(data1))

KNN_IMPUTED_DATA2.head()
```

```python
sns.heatmap(KNN_IMPUTED_DATA2.isnull(),cmap='Oranges')
plt.show()
```

# Applying the Imbalancing technique  NEAR MISS

```python
import numpy as np
import pandas as pd
from collections import Counter
from imblearn.under_sampling import NearMiss
```

```python
# titanic 

x = KNN_IMPUTED_DATA1.drop('Survived', axis=1)
y = KNN_IMPUTED_DATA1['Survived']
```

```python
# DATA.csv

x1 = KNN_IMPUTED_DATA2.drop('target', axis=1)
y1 = KNN_IMPUTED_DATA2['target']
```

# Version 1

```python
nearmiss = NearMiss(n_neighbors=2)

x_nm ,y_nm   = nearmiss.fit_resample(x,y)


x = pd.DataFrame(x_nm, columns=data.cols[::-1])
y = pd.Series(y_nm, name='Survived')

data = pd.concat([x, y], axis=1)
```

```python
nearmiss = NearMiss(n_neighbors=2)

x1_nm ,y1_nm   = nearmiss.fit_resample(x1,y1)


x1 = pd.DataFrame(x1_nm, columns=data1.columns[::-1])
y1 = pd.Series(y1_nm, name='target')

data = pd.concat([x1, y1], axis=1)
```
