# **Usage / Implementation of Pandas**

# Reading the csv file

```python
import pandas as pd

df = pd.read_csv('Student.csv')

df
```

```python
# print head top three
df.head(3)
```

```python
# show column names and data types
df.dtypes
```

```python
# show the summaries of the data
df.describe()
```

```python
# iformation of the data
df.info()
```

```python
# display average score in science
df['Science'].mean()
```

```python
# adding a new column total which is sum of math ,science,english
df['Total'] = df['Math'] + df['Science'] + df['English']

df
```

```python
# add another column result with value pass if total >= 240 , else fail
df['Result'] = df['Total'].apply(lambda x: 'Pass' if x>=240 else 'Fail')

df
```

```python
# showing how many students passed and failed
df['Result'].value_counts()
```

```python
# filtering only females students who scored more than 80 in math
df[(df['Gender'] == 'Female') & (df['Math'] > 80)]
```

```python
# simple bar plot of total score
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
plt.title('Total Score of Students')
plt.bar(df.Name, df['Total'], color='blue')
plt.xlabel('Student Name') 
plt.show()

```

```python
# pie Chart of result
df['Result'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, shadow=True, figsize=(8, 8))
plt.title('Result of Students')
plt.show()
```

```python
# histogram of math score
# df['Math'].plot.hist(bins=10, alpha=0.7, color='blue')
plt.hist(df['Math'], bins=10, alpha=0.9, color='blue')
plt.title('Math Score Distribution')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```
