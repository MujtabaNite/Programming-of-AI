# **Using Nested dictionary and Applying the imputer technique**

# define the data

```python
import numpy as np
import pandas as pd


data = {
    'Name': ['awais', 'fazil', 'hadain','jabbar'],
    'Age': [25, np.nan, 35,np.nan],
    'City': ['Lahore', 'Karachi', 'Rawalpindi','Islamabad'],
    'Scores': {
        'Math': [90, 85, np.nan, 95],
        'Science': [92, np.nan, 85,np.nan],
        'English': [88, 90, 87,np.nan]
    }
}

_data = {
    'Name': data['Name'],
    'Age': data['Age'],
    'City': data['City'],
    'Math': data['Scores']['Math'],
    'Science': data['Scores']['Science'],
    'English': data['Scores']['English']
}

df = pd.DataFrame(_data)
print(df)
```

```python
df_Simple = df_KNN = df_itreative = df.copy()
```

```python
# applying simple imputer 
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
# Impute missing values in the 'Age' and 'Scores' columns
df_Simple[['Age', 'Math', 'Science','English']] = imputer.fit_transform(df_Simple[['Age', 'Math', 'Science','English']])
# Display the DataFrame after imputation
print("\nDataFrame after imputation:")
print(df_Simple)
```

```python
# applying the knn imputer
from sklearn.impute import KNNImputer
knn_imputer = KNNImputer(n_neighbors=2)
# Impute missing values using KNN
df_KNN[['Age', 'Math', 'Science','English']] = knn_imputer.fit_transform(df_KNN[['Age', 'Math', 'Science','English']])
# Display the DataFrame after KNN imputation
print("\nDataFrame after KNN imputation:")
print(df_KNN)
```

```python
# applying the itreative imputer
from sklearn.experimental import enable_iterative_imputer  # noqa
from sklearn.impute import IterativeImputer

iterative_imputer = IterativeImputer(max_iter=10, random_state=0)
# Impute missing values using Iterative Imputer
df_itreative[['Age', 'Math', 'Science','English']] = iterative_imputer.fit_transform(df_itreative[['Age', 'Math', 'Science','English']])
# Display the DataFrame after Iterative Imputation
print("\nDataFrame after Iterative Imputation:")
print(df_itreative)
```
