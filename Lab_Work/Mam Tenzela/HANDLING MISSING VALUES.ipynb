# **Handling Missing Values**

* Implement methods to identify using  isna() or isnull() funcitons visulize the missing data using heatmap or barchart

```python
import pandas as pd

data = pd.read_csv('titanic_toy.csv')


data.head()
```

```python
data.info()
```

```python
data_using_ISNA = data.isna()
data_using_ISNA
```

```python
data_using_isNull = data.isnull()
data_using_isNull.head()
```

```python
# visualizing the missing values  using heatmap
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
sns.heatmap(data_using_ISNA, cbar=False, cmap ='viridis')
plt.title('Missing Values Heatmap')
plt.show()

```

```python
# Visualizing the missing values using bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=data.columns, y=data.isna().sum())
plt.title('Missing Values Count')
plt.show()
```
