# ALGORITHM 

titanic_toy.shape  # i , j   shape[0]  and shape[1]   tuple

* loop one on row  
* loop two on column 
* get loc on i j 
* display 

```python
#  printing  column number and row number of the missing values in titanic_toy
for i in range(titanic_toy.shape[0]):
    for j in range(titanic_toy.shape[1]):
        if pd.isnull(titanic_toy.iloc[i, j]):
            print(f"Row: {i}, Column: {j}")

```

```python
#  printing  column number and row number of the missing values in depression_dataset
for i in range(depression_dataset.shape[0]):
    for j in range(depression_dataset.shape[1]):
        if pd.isnull(depression_dataset.iloc[i, j]):
            print(f"Row: {i}, Column: {j}")
```

```python
#  printing  column number and row number of the missing values in passenger
for i in range(Passenger.shape[0]):
    for j in range(Passenger.shape[1]):
        if pd.isnull(Passenger.iloc[i, j]):
            print(f"Row: {i}, Column: {j}")
```

```python
# disaplyint the missing values in titanic_toy using seaborn heat map

import seaborn as sns 
import matplotlib.pyplot as plt

sns.heatmap(titanic_toy.isnull(), cbar=False)
plt.title("Missing values in titanic_toy")
plt.show()
```

```python
# displayint the missing values in depression_dataset using seaborn heat map

sns.heatmap(depression_dataset.isnull(), cbar=False)
plt.title("Missing values in depression_dataset")
plt.show()
```

```python
# displayint the missing values in Passenger using seaborn heat map

sns.heatmap(Passenger.isnull(), cbar=False)
plt.title("Missing values in Passenger")
plt.show()
```
