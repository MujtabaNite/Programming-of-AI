```python
# making a recurive function to calculate the average student grade for each student and display the
# name along with average grade   the function will take a dictionary as an argument and 
#  index from where you want to start the calculation then caluate the average and display the name and average grade

def calculate_average_grade(students, index=0):
    if index >= len(students):
        return
    student_id = list(students.keys())[index]
    student = students[student_id]
    name = student["Name"]
    grades = student["course"].values()
    average_grade = sum(grades) / len(grades)
    print(f"Student ID: {student_id}, Name: {name}, Average Grade: {average_grade:.2f}")
    calculate_average_grade(students, index + 1)


calculate_average_grade(students)
```
